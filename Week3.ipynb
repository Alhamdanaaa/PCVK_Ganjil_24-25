2. Implementasikan transformasi contrast pada Google Colaboratory menggunakan
formula yang terdapat pada bagian Ulasan Teori untuk kontras, sehingga
menghasilkan keluaran seperti berikut:

```python
print('Mengubah kontras dan tingkat kecerahan citra')
print('--------------------------------------------')
try:
  brightness = int(input("Masukkan tingkat kecerahan [-255 - 255]: "))
  contras = int(input("Masukkan kontras [1.0 - 3.0]: "))
except ValueError:
  print("Error, not a number")

original = cv.imread('/content/drive/MyDrive/PCVK/female.tiff')
result_image = np.zeros(original.shape,original.dtype)

#akses per piksel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      result_image[y,x,c] = np.clip(contras * (original[y,x,c]) + brightness,-255,255)

final_frame = cv.hconcat((original, result_image))
cv2_imshow(final_frame)
```

3. Implementasikan transformasi logarithmic brightness pada Google Colaboratory
menggunakan formula yang terdapat pada bagian Ulasan Teori untuk transformasi
log, sehingga menghasilkan keluaran seperti berikut:

```python
import math

print('Mengubah tingkat kecerahan citra dengan Transformasi Log')
print('--------------------------------------------------------')

try:
  brightness = int(input("Masukkan tingkat kecerahan [-255 - 255]: "))
except ValueError:
  print("Error, not a number")

original = cv.imread('/content/drive/MyDrive/PCVK/female.tiff')
result_image = np.zeros(original.shape,original.dtype)

#akses per piksel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      result_image[y,x,c] = np.clip(brightness * math.log(1 + original[y,x,c]),-255,255)

final_frame = cv.hconcat((original, result_image))
cv2_imshow(final_frame)
```

5. Tampilkanlah warna tertentu pada citra, dan ubah warna lain menjadi grayscale. Misal,
tampilkan warna biru pada citra masukan dan ubah bagian lain yang tidak berwarna
biru menjadi grayscale seperti pada contoh berikut:

```python
original = cv.imread('/content/drive/MyDrive/PCVK/ktp4.png')
original = cv.cvtColor(original,cv.COLOR_BGR2RGB)
result_image = np.zeros(original.shape,original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    R,G,B = original[y,x]

    #Memeriksa apakah piksel berwarna biru atau warna merah
    if (B > 0 and R < 100 and G < 200) or (B > 50 and R > 0 and G < 200):
        # Jika biru atau merah, gunakan warna asli
        result_image[y, x] = original[y, x]
    else:
        # Jika bukan biru atau warna merah, ubah menjadi grayscale
        gray_value = 0.21 * R + 0.72 * G + 0.07 * B
        result_image[y, x] = [gray_value, gray_value, gray_value]

final_frame = plt.subplot(1,2,1)
plt.imshow(original)
plt.subplot(1,2,2)
plt.imshow(result_image)
plt.show()
```
