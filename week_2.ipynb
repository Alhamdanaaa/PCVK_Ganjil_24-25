2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!


```python
image = cv.imread('/content/drive/MyDrive/PCVK/fox.jpg')

# Mengkonversi gambar dari BGR ke RGB
imgTugas = cv.cvtColor(image, cv.COLOR_BGR2RGB)

R, G, B = cv.split(imgTugas)
image_RB = cv.merge([R, np.zeros_like(G),B])
image_GB = cv.merge([np.zeros_like(R),G,B])

final_image = cv.hconcat((image_RB,image_GB))
cv2_imshow(final_image)
```
