```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import glob
from math import log10, sqrt
```

1. Buat Gamma Correction sesuai dengan petunjuk

```python
print(' Gamma Corection pada citra ')
print('----------------------------')
try:
  gamma = int(input('Masukkan nilai Gamma: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/PCVK/female.tiff')
# original = cv.cvtColor(original, cv.COLOR_BGR2RGB)
result_image = np.zeros(original.shape,original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    result_image[y,x] = 255*(original[y,x]/255)**(1/gamma)

final_frame = cv.hconcat((original, result_image))
cv2_imshow(final_frame)
```

2. Buat Simulasi Image Depth

```python
bit_depth=2
level = 255 / (pow(2,bit_depth)-1)
original = cv.imread('/content/drive/MyDrive/PCVK/female.tiff', cv.IMREAD_GRAYSCALE)
depth_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    depth_image[y,x] = round(original[y,x]/level)*level

final_frame = cv.hconcat((original,depth_image))
cv2_imshow(final_frame)
```

3. Buat modul Average Denoising

```python
def PSNR(img1, img2):
  mse = np.mean((img1 - img2) ** 2)
  if(mse == 0): # MSE 0 maka tidak ada noise sama sekali, sehingga PSNR tidak memiliki arti
    return 100
  max_pixel = 255.0
  psnr = 20 * log10(max_pixel / sqrt(mse))
  return psnr
```

*   Menggunakan 5 citra


```python
# Baca citra asli
original_image = cv.imread('/content/drive/MyDrive/PCVK/galaxy.jpg')
# Baca semua citra dengan Gaussian Noise
cv_img =[]
for img in glob.glob('/content/drive/MyDrive/PCVK/noises/*.jpg'):
  n = cv.imread(img)
  cv_img.append(n)

average_image = np.zeros_like(cv_img[0], dtype=np.float64)
total_img = 5
for noisy_img in range(total_img):
  average_image += cv_img[noisy_img]

average_image /= total_img
cv2_imshow(average_image)

psnr_value = PSNR(original_image, average_image)
print(f"PSNR antara citra asli dan citra denoised: {psnr_value}")
```

5. Lakukan percobaan menggunakan operator lain dan tunjukkan hasilnya pada modul ini.



*   NOT (komplemen)



```python
fox = cv.imread('/content/drive/MyDrive/PCVK/fox.jpg')
fox = cv.resize(fox,(fox.shape[1]//2,fox.shape[0]//2))
image = np.zeros(fox.shape,fox.dtype)

cv.rectangle(image,pt1=(80,80),pt2=(190,190),color=(255,255,255),thickness=-1)
result_image = cv.bitwise_not(fox)

final_image = cv.hconcat((fox,result_image))
cv2_imshow(final_image)
```



*   OR (Atau)




```python
fox = cv.imread('/content/drive/MyDrive/PCVK/fox.jpg')
fox = cv.resize(fox,(fox.shape[1]//2,fox.shape[0]//2))
image = np.zeros(fox.shape,fox.dtype)

cv.rectangle(image,pt1=(80,80),pt2=(190,190),color=(255,255,255),thickness=-1)
result_image = image | fox

final_image = cv.hconcat((fox,result_image))
cv2_imshow(final_image)
```



*   AND (Dan)


```python
fox = cv.imread('/content/drive/MyDrive/PCVK/fox.jpg')
fox = cv.resize(fox,(fox.shape[1]//2,fox.shape[0]//2))
image = np.zeros(fox.shape,fox.dtype)

cv.rectangle(image,pt1=(80,80),pt2=(190,190),color=(255,255,255),thickness=-1)
result_image = image & fox

final_image = cv.hconcat((fox,result_image))
cv2_imshow(final_image)
```



*   NAND (Not
And)



```python
fox = cv.imread('/content/drive/MyDrive/PCVK/fox.jpg')
fox = cv.resize(fox,(fox.shape[1]//2,fox.shape[0]//2))
image = np.zeros(fox.shape,fox.dtype)

cv.rectangle(image,pt1=(80,80),pt2=(190,190),color=(255,255,255),thickness=-1)
result_image = cv.bitwise_not(cv.bitwise_and(fox,image))

final_image = cv.hconcat((fox,result_image))
cv2_imshow(final_image)
```
